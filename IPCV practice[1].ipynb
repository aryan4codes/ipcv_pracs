```python
import matplotlib.pyplot as plt

def compare(input_image, output_image, title):
    # Create a figure with specified size
    fig = plt.figure(figsize=(10, 8))
    
    # Add the original image as the first subplot
    fig.add_subplot(2, 2, 1)
    plt.imshow(input_image, cmap='gray')
    plt.title("Original image")
    
    # Add the output image as the second subplot
    fig.add_subplot(2, 2, 2)
    plt.imshow(output_image, cmap='gray')
    plt.title(title)
    
    # Display the figure
    plt.show()
```

```python
def erosion(image, s):
    # Ensure the image is binary (values should be either 0 or 1)
    image = (image >= 1).astype(np.uint8)
    
    # Create an empty image for the result
    new_image = np.zeros_like(image)
    
    n, m = image.shape
    elem_h, elem_w = s.shape
    
    pad_h, pad_w = elem_h // 2, elem_w // 2
    
    # Pad the image to handle borders
    padded_img = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant', constant_values=0)
    
    # Perform the erosion operation
    for i in range(n):
        for j in range(m):
            # Extract the neighbourhood region corresponding to the kernel size
            neighbourhood = padded_img[i:i + elem_h, j:j + elem_w]
            
            # Check if the kernel fits in the neighbourhood
            if np.all(neighbourhood[s == 1] == 1):
                new_image[i, j] = 1  # Set pixel to 1 if the kernel fits
    
    return new_image
```

```python
def dilate(image,kernel):
    image = (image >= 1).astype(np.uint8)
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape
    
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2
    
    padded_image = np.pad(image,((pad_height,pad_height),(pad_width,pad_width)))
    dilated_image = np.zeros_like(image)
    
    for i in range(image_height):
        for j in range(image_width):
            region = padded_image[i:i+kernel_height,j:j+kernel_width]
            if np.any(region[kernel==1]==1):
                dilated_image[i,j] = 1
            else:
                dilated_image[i,j] = 0
                
    return dilated_image
```

```python
def opening(image,kernel):
    image = (image >= 1).astype(np.uint8)
    eroded_image = erode(image,kernel)
    opened_image = dilate(eroded_image,kernel)
    return opened_image
```

```python
def closing(image,kernel):
    image = (image >= 1).astype(np.uint8)
    dilated_image = dilate(image,kernel)
    closed_image = erode(dilated_image,kernel)
    return closed_image
```

```python
def hit_or_miss(image,fg_kernel,bg_kernel):
    c_img = 1 - image
    eroded_fg = erode(image,fg_kernel)
    eroded_bg = erode(c_img,bg_kernel)
    output = np.logical_and(eroded_fg,eroded_bg).astype(np.uint8)
    return output
```

```python
#Exp6 dealing with filters

def apply_filter(image,kernel):
    h,w = image.shape
    kernel_size = kernel.shape[0]
    pad = kernel_size // 2
    padded_image = np.pad(image,pad,mode='constant',constant_values=0)
    result = np.zeros_like(image)
    
    for i in range(h):
        for j in range(w):
            region = padded_image[i:i+kernel_size,j:j+kernel_size]
            result[i,j] = np.sum(region * kernel)
    
    return result.astype(np.uint8)        
```

```python
def high_boost_filter(image, A):
    # High-boost kernel
    kernel = np.array([
        [-1, -1, -1],
        [-1, (9 * A - 1), -1],
        [-1, -1, -1]
    ], dtype=np.float32)
    
    hb_image = apply_filter(image, kernel)
    hb_image = np.clip(hb_image, 0, 255).astype(np.uint8)
    return hb_image

hb_image = high_boost_filter(six, A)

compare(six, hb_image, f"High Boost Filter {A}")
```

```python
# log transformation
def log_tranform(image):
    if np.max(image)==0:
        return image
    c = 255 / np.log(1 + np.max(image))
    result = c * (np.log(1 + image))
    return result.astype(np.uint8)
```

```python
def min_max_stretching(image):
    mini = np.min(image)
    maxi = np.max(image)
    result = (image - mini) * 255  / (maxi - mini)
    return result.astype(np.uint8)
```

```python
min_max = min_max_stretching(four)
compare(four,min_max,"Contrast stretching")
```

```python
def contrast_stretching_slopes(image,r1,r2,s1,s2):
    image = image.astype(float)
    h,w = image.shape
    slope1 = s1/r1
    slope2 = (s2 - s1) / (r2 - r1)
    slope3 = (255 - s2) / (255 - r2)
    result = np.zeros_like(image)
    for i in range(h):
        for j in range(w):
            p = image[i,j]
            if p < r1:
                result[i,j] = slope1 * p
            elif (p >= r1 and p < r2):
                result[i,j] = slope2 * (p - r1) + s1
            elif (p >= r2):
                result[i,j] = slope3 * (p - r2) + s2
                
    result = np.clip(result,0,255).astype(np.uint8)
    return result
```

```python
css = contrast_stretching_slopes(four,10,20,15,20)
compare(four,css, "Contrast stretching with slopes")
```

```python
def histogram_eq(image):
    hist,_ = np.histogram(image.flatten(),bins=256,range=[0,256])
    cdf = hist.cumsum()
    cdf_normalised = (cdf - cdf.min()) / (cdf.max() - cdf.min()) * 255 
    equalised_image = cdf_normalised[image].astype(np.uint8)
    return equalised_image
```

```python
he = histogram_eq(four)
compare(four,he,"Histogram equalisation")
```

```python
def add_gaussian_noise(image,mean,std):
    noise = np.random.normal(mean,std,image.shape)
    noisy_image = image + noise
    noisy_image = np.clip(image,0,255).astype(np.uint8)
    return noisy_image
adn = add_gaussian_noise(four,15,5)
compare(four,adn,"Gaussian noise")
```

```python
def salt_and_pepper_noise(image,salt_prob=0.02, pepper_prob=0.02):
    h,w = image.shape
    noisy_image = np.copy(image)
    
    salt_mask = np.random.rand(h,w) < salt_prob
    noisy_image[salt_mask] = 255
    
    pepper_mask = np.random.rand(h,w) < pepper_prob
    noisy_image[pepper_mask] = 0
    
    return noisy_image

spn = salt_and_pepper_noise(four)
compare(four,spn,"salt and pepper noise")
```

```python
def median_filter(image,kernel_size):
    h,w = image.shape
    pad_size = kernel_size // 2
    padded_image = np.pad(image,((pad_size,pad_size),(pad_size,pad_size)),mode='edge')
    result = np.zeros_like(image)
    
    for i in range(h):
        for j in range(w):
            window = padded_image[i:i+kernel_size,j:j+kernel_size]
            result[i,j] = np.median(window)
    return result

med_f = median_filter(four,4)
compare(four,med_f,"Median filter")
```

```python
def avg_filter(image,kernel_size):
    h,w = image.shape
    pad_size = kernel_size // 2
    padded_image = np.pad(image,((pad_size,pad_size),(pad_size,pad_size)),mode='edge')
    result = np.zeros_like(image)
    
    for i in range(h):
        for j in range(w):
            window = padded_image[i:i+kernel_size,j:j+kernel_size]
            result[i,j] = np.mean(window)
    return result

avg_f = avg_filter(four,3)
compare(four,avg_f,"Average filter")
```

```python
def apply_transform(image,filter_mask):
    dft = fft2(image) 
    dft_shift = fftshift(dft)
    filtered_dft = dft_shift * filter_mask
    ifft_shift = ifftshift(filtered_dft)
    filtered_img = ifft2(ifft_shift)
    return np.abs(filtered_img)
```

```python
import numpy as np
from numpy.fft import fft2, fftshift, ifft2, ifftshift




def apply_transform(image, filter_mask):


    # Step 1: Compute the 2D FFT of the image (frequency domain)
    dft = fft2(image) 
    
    # Step 2: Shift zero frequency to the center of the spectrum
    dft_shift = fftshift(dft)
    
    # Step 3: Apply the filter (element-wise multiplication)
    filtered_dft = dft_shift * filter_mask
    
    # Step 4: Shift the filtered DFT back to original position
    ifft_shift = ifftshift(filtered_dft)
    
    # Step 5: Apply inverse FFT to get the filtered image in the spatial domain
    filtered_img = ifft2(ifft_shift)
    
    # Step 6: Return the magnitude (real part) of the filtered image
    return np.abs(filtered_img)
```

```python
def ideal_low_pass_filter(image,cutoff):
    
    rows, cols = image.shape
    u = np.fft.fftfreq(cols, 1.0)
    v = np.fft.fftfreq(rows, 1.0)
    U,V = np.meshgrid(u,v)
    D = np.sqrt(U**2 + V**2)
    
    
    H = np.zeros_like(image)
    H[D < cutoff] = 1
    idlf = apply_filter(image,H)
    return idlf

idlpf = ideal_low_pass_filter(six,0.15)
compare(six,idlpf,"Ideal low pass filter")
```

```python
def ideal_high_pass_filter(image,cutoff):
    return 1 - ideal_low_pass_filter(image,cutoff)

idhpf = ideal_low_pass_filter(six,0.85)
compare(six,idhpf,"Ideal high pass filter")
```

```python
def gaussian_low_pass_filter(image,cutoff):
    rows, cols = image.shape
    u = np.fft.fftfreq(cols, 1.0)
    v = np.fft.fftfreq(rows, 1.0)
    U,V = np.meshgrid(u,v)
    D = np.sqrt(U**2 + V**2)
    
    
    H = np.exp(-D**2 / (2 * (cutoff**2))) 
    glpf=apply_filter(image,H)
    
    return glpf

glpf_result = gaussian_low_pass_filter(four,0.07)
compare(four,glpf_result,"Gaussian low pass filter")
```

```python
def gaussian_high_pass_filter(image,cutoff):
    return 1 - gaussian_low_pass_filter(image,cutoff)

ghpf_result = gaussian_high_pass_filter(four, 0.03)
compare(four, ghpf_result, "Gaussian High Pass Filter")
```

```python
def butterworth_low_pass_filter(image, cutoff, order=2):
    rows, cols = image.shape
    u = np.fft.fftfreq(cols, 1.0)
    v = np.fft.fftfreq(rows, 1.0)
    U, V = np.meshgrid(u, v)
    D = np.sqrt(U**2 + V**2)  # Distance from the center in frequency domain

    # Butterworth filter transfer function
    H = 1 / (1 + (D / cutoff)**(2 * order))
    blpf = apply_filter(image, H)  # Apply the filter to the image
    return blpf

# Butterworth High-Pass Filter (BHPF)
def butterworth_high_pass_filter(image, cutoff, order=2):
    rows, cols = image.shape
    u = np.fft.fftfreq(cols, 1.0)
    v = np.fft.fftfreq(rows, 1.0)
    U, V = np.meshgrid(u, v)
    D = np.sqrt(U**2 + V**2)  # Distance from the center in frequency domain

    # Butterworth high-pass filter transfer function
    H = 1 / (1 + (cutoff / D)**(2 * order))
    bhpf = apply_filter(image, H)  # Apply the filter to the image
    return bhpf

# Example usage with cutoff frequency and order
cutoff_frequency = 0.1  # Cutoff frequency for the filter
order = 2  # Order of the Butterworth filter (controls the sharpness)

# Apply Butterworth Low-Pass Filter (BLPF)
blpf_result = butterworth_low_pass_filter(four, cutoff_frequency, order)
compare(four, blpf_result, "Butterworth Low Pass Filter")

# Apply Butterworth High-Pass Filter (BHPF)
bhpf_result = butterworth_high_pass_filter(four, cutoff_frequency, order)
compare(four, bhpf_result, "Butterworth High Pass Filter")
```

```python
def region_growing(img, thresh):
    r, c = img.shape
    seed = (random.randint(0, r - 1), random.randint(0, c - 1))

    seg = np.zeros_like(img, dtype=np.uint8)
    seed_x, seed_y = seed
    seg[seed_x, seed_y] = 255
    region_mean = img[seed_x, seed_y]

    growing = True
    pixels = [(seed_x, seed_y)]

    while growing:
        new_pixels = []
        for x, y in pixels:
            for dx, dy in [(1, 0), (-1, 0), (0, 1), (0, -1)]:
                nx, ny = x + dx, y + dy

                if 0 <= nx < r and 0 <= ny < c and seg[nx, ny] == 0:
                    if abs(int(img[nx, ny]) - region_mean) < thresh:
                        seg[nx, ny] = 255
                        new_pixels.append((nx, ny))

        if new_pixels:
            region_mean = np.mean([img[x, y] for x, y in new_pixels])
            pixels = new_pixels
        else:
            growing = False

    return seg
```

```python
result = region_growing(six,60)
compare(six, result, 'Region Growing Segmentation')
```

```python
def merge_regions(regions):
    while True:
        merged = False
        new_regions = []
        while regions:
            current = regions.pop()
            was_merged = False

            for idx, region in enumerate(new_regions):
                if abs(np.mean(region) - np.mean(current)) < 5:
                    new_regions[idx] = np.vstack([region, current])
                    was_merged = True
                    merged = True
                    break

            if not was_merged:
                new_regions.append(current)

        regions = new_regions
        if not merged:
            break
    return regions
```

```python
def split_and_merge(image, num_regions):
    r, c = image.shape
    step = r // num_regions

    regions = [np.arange(i, min(i + step, r)) for i in range(0, r, step)]
    regions = merge_regions(regions)

    output_image = np.zeros_like(image, dtype=np.uint8)
    for region in regions:
        for row in region:
            output_image[row, :] = ((np.mean(image[region, :]) - image[row, :]) < 10) * 255

    return output_image
```

```python
s = split_and_merge(six, 2)
compare(six, s, 'Split and Merge Segmentation')
```
